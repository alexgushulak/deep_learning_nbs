```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import time
from torchvision import transforms, datasets
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
```

```python
train = datasets.MNIST("", train=True, download=True,
                       transform = transforms.Compose([transforms.ToTensor()]))

test = datasets.MNIST("", train=False, download=True,
                      transform = transforms.Compose([transforms.ToTensor()]))
```

```python
trainset = torch.utils.data.DataLoader(train, batch_size=64, shuffle=True)
testset = torch.utils.data.DataLoader(test, batch_size=64, shuffle=False)
```

```python
class Net(nn.Module):
  def __init__(self):
    super().__init__()
    self.sz = 256
    self.fc1 = nn.Linear(28*28, self.sz) # first fully connected layer2
    self.fc2 = nn.Linear(self.sz, self.sz)
    self.fc3 = nn.Linear(self.sz, self.sz)
    self.fc4 = nn.Linear(self.sz, 10)

  def forward(self, x):
    x = F.hardtanh(self.fc1(x))
    x = F.hardtanh(self.fc2(x))
    x = F.hardtanh(self.fc3(x))
    x = self.fc4(x)

    return F.log_softmax(x, dim=1)

my_model = Net()
```

```python
optimizer = optim.SGD(my_model.parameters(), lr=1e-4)
EPOCHS = 1
q = 0
for epoch in range(EPOCHS):
  for data in trainset:
    if q % 100 == 0:
      print(q)
    q += 1
    X, y = data
    my_model.zero_grad()
    output = my_model(X.view(-1, 28*28))
    loss = F.nll_loss(output, y)
    loss.backward()
    optimizer.step()
    print(loss)
```

```python
correct, total = 0,0

with torch.no_grad():
  for data in testset:
    X,y = data
    output = my_model(X.view(-1, 784))
    for idx, i in enumerate(output):
      if torch.argmax(i) == y[idx]:
        correct += 1
      total += 1

print("Accuracy: ", round(correct/total, 3)*100, "%")
```
